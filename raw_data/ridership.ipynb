# Peak or non-peak

```python
df2= pd.read_excel('Additional Platform Time.xlsx')
df2
```

```python
df2.month = pd.to_datetime(df2.month, format="%Y-%m")
```

```python
df2
```

```python
passengers = df2.groupby(['period', 'month']).num_passengers.mean().to_frame()
passengers
```

```python
passengers.reset_index(inplace=True)
```

```python
passengers
```

## Offpeak

```python
offpeak = passengers[passengers.period=='offpeak']
offpeak.tail(24)
```

```python
offpeak.tail()
```

```python
(
    ggplot(offpeak, aes('month', 'num_passengers')) +
           geom_col(color="red", stat="identity") +
           theme(figure_size=(16,8)) +
           labs(x= "Date", y="Number of passengers", title="Offpeak")
)
```

## Peak

```python
peak = passengers[passengers.period=='peak']
peak.head(12)
```

```python
peak.tail(24)
```

```python
(
    ggplot(peak, aes('month', 'num_passengers')) +
           geom_col(color="red", stat="identity") +
           theme(figure_size=(16,8)) +
           labs(x= "Date", y="Number of passengers", title="Peak")
)
```

```python
df3 = peak.merge(offpeak, on='month')
df3
```

```python
df3.columns = ['period_1', 'date', 'peak', 'period_2', 'offpeak']
df3
```

```python
df3 = df3.drop(['period_1', 'period_2'], axis=1)
df3
```

```python
peak2 = df3.resample('Y', on='date').peak.mean().to_frame()
peak2
```

```python
offpeak2 = df3.resample('Y', on='date').offpeak.mean().to_frame()
offpeak2
```

```python
df4 = peak2.merge(offpeak2, on='date')
df4
```

```python
df4['gap'] = df4.peak - df4.offpeak
df4
```

```python
df4.to_csv('peak.csv')
```
